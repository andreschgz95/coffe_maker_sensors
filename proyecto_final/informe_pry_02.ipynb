## Premiación

Depende del puntaje obtenido, y bajo criterio del instructor será entregado un premio, **a escoger** entre:

- PyBadge: Especial para juegos, Makecode, Circuitpython
- ESP32 Cam: Wifi, cámara y visión computarizada incipiente
- Maix Bit: Alta velocidad, inteligencia artificial, visión computarizada y wifi

### PyBadge
Microcontroladora 32 bits de mediana carga de procesamiento, con pantalla, botones para videojuego y cargador de batería.

![PyBadge](https://cdn2.bigcommerce.com/server3700/64749/products/1600/images/5084/3939-00__34326.1559755324.1280.1280.jpg?c=2)

## ESP32 Cam
Las ventajas conocidas de las ESP32 pero en un formato pequeño más conector de batería.

**Importante**: Requiere tarjeta adicional para programar.

![ESP32 Cam](https://cdn2.bigcommerce.com/server3700/64749/products/1588/images/5046/ESP32-CAM__26233.1558704012.1280.1280.png?c=2)

### Maix Bit
Microcontroladora de 64 bits, 2 nucleos RISCV a 400MHz, Acelerador para redes neuronales, visión computarizada y escucha computarizada, incluye ESP8265 para red Wifi.

![Maix Bit](https://www.seeedstudio.site/media/catalog/product/cache/134ea8534034ded9d909870d8862ea94/b/i/bit-front.jpg)
